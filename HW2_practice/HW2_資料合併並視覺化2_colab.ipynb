```python
#以工作階級為groupby，計算出各個工作階級中工作時數的平均數
hours_per_week_by_workclass = data1.groupby(['workclass'])['hours_per_week'].agg(['mean'])


# 保留 'column_name' 欄位在新的資料中
grouped_data_with_column = hours_per_week_by_workclass.reset_index()

# 將dataframe中第一行的資料刪除
grouped_data_with_column.drop(0, inplace=True)

grouped_data_with_column
```

```python
# 以性別進行 groupby 找出教育年限的分布
sex_education_num = data1.groupby('sex')['education_num'].agg(['mean', 'std', 'median'])

sex_education_num
```

```python
# 不同的性別教育年限分布會不會有差別?

sex_education_num = data1[['sex', 'education_num']]
import plotly.express as px

fig = px.violin(sex_education_num, y="education_num", x="sex", color="sex",
                box=True, points="all",
          hover_data=sex_education_num.columns)
fig.show()
```

```python
# 以種族進行groupby 找出教育年限的分布
race_education_num = data1.groupby('race')['education_num'].agg(['mean', 'std' , 'median'])
race_education_num
```

```python
#不同種族是否會影響到教育年限的分佈？
race_education_num = data1[['race', 'education_num']]

import plotly.express as px

fig = px.violin(race_education_num, y="education_num", x="race", color="race",
                box=True, points="all",
          hover_data=race_education_num.columns)
fig.show()

#以資料呈現樣態相似度來看，白人、亞太裔為一組。而黑人、美洲印地安和愛斯基摩人為一組。
#其中，白人以及亞太裔的組別25%的人教育年限在9歲左右，50%的人集中在10歲，75%的人集中在13歲
#而黑人、美洲印地安和愛斯基摩人的組別25%集中在9歲左右，50%的人也是在9歲附近，75%的人則是在10歲。
#整體來看，前面組別的資料分佈較為鬆散，而後面組別的資料較為集中。
```

```python
#以教育程度為groupby，計算出不同教育程度中高收入以及低收入的人數
income_by_education = data1.groupby(['education','income'])['sex'].count().unstack()

income_by_education
```

```python
# 計算每個教育程度中，高收入與低收入人數佔該教育程度總人數的百分率 (測試錯誤)
income_by_education['income_ratio（high）'] =income_by_education[' >50K'] / income_by_education.sum(axis=1)*100
income_by_education['income_ration（low）'] = income_by_education[' <=50K'] / income_by_education.sum(axis=1)*100

income_by_education
```

```python
# 教育程度的不同是否會影響到高收入與低收入的佔比?
# 在這裡出圖出現錯誤是因在income_by_education這個dataframe中10th、11th、12th這些欄位其實都是欄位名稱而不是欄位資料
fig = px.bar(income_by_education, x='education', y=['income_ratio（high）', 'income_ration（low）'], barmode='group')

#配置圖表
fig.update_layout(
    title='各個教育程度中高收入與低收入人數佔各教育程度人數的百分比',
    xaxis_title='教育程度',
    yaxis_title='百分比'
)
fig.show()
```

```python
# 計算每個教育程度中，高收入與低收入人數佔該教育程度總人數的百分率 (正確的)
income_by_education['income_ratio（high）'] =income_by_education[' >50K'] / income_by_education.sum(axis=1)*100
income_by_education['income_ration（low）'] = income_by_education[' <=50K'] / income_by_education.sum(axis=1)*100

# 保留 'column_name' 欄位在新的資料中(將Federal-gov、Local-gov、Never-worked這些改成欄位資料)
grouped2_data_with_column = income_by_education.reset_index()

grouped2_data_with_column

```
