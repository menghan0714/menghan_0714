```python
data1.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
              'marital_status', 'occupation', 'relationship', 'race', 'sex',
              'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
```

```python
#groupby
hours_per_week_by_workclass = data1.groupby(['workclass'])['hours_per_week'].agg(['mean'])


#  'column_name' 
grouped_data_with_column = hours_per_week_by_workclass.reset_index()

# dataframe
grouped_data_with_column.drop(0, inplace=True)

grouped_data_with_column
```

```python
#  groupby 
sex_education_num = data1.groupby('sex')['education_num'].agg(['mean', 'std', 'median'])

sex_education_num
```

```python
# ?

sex_education_num = data1[['sex', 'education_num']]
import plotly.express as px

fig = px.violin(sex_education_num, y="education_num", x="sex", color="sex",
                box=True, points="all",
          hover_data=sex_education_num.columns)
fig.show()
```

```python
#groupby
income_by_education = data1.groupby(['education','income'])['sex'].count().unstack()

income_by_education
```

```python
#  ()
income_by_education['income_ratiohigh'] =income_by_education[' >50K'] / income_by_education.sum(axis=1)*100
income_by_education['income_rationlow'] = income_by_education[' <=50K'] / income_by_education.sum(axis=1)*100

income_by_education
```

```python
# ?
# income_by_educationdataframe10th11th12th
fig = px.bar(income_by_education, x='education', y=['income_ratiohigh', 'income_rationlow'], barmode='group')

#
fig.update_layout(
    title='',
    xaxis_title='',
    yaxis_title=''
)
fig.show()
```
