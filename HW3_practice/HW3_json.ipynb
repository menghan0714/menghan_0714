```python
# 想要把結果轉換成像json一樣的槽狀結構，因此把結果轉換成json中的字符串格式

import json

file_path = '/content/drive/My Drive/practice/Tainan.json'

with open(file_path, 'r') as f:
    data = json.load(f)

pretty_json = json.dumps(data, indent=4)
print(pretty_json)
```

```python
# 詢問chatGPT，上面的結果是經過Unicode編碼的結果。而這樣的結果很常發生在列印元素中包含並非是ASCII字符的字符串時(例:中文)
# 因此使用ensure_ascii=False參數確保輸出的不是Unicode的轉義序列而是可讀的中文文字符

import json

file_path = '/content/drive/My Drive/practice/Tainan.json'

with open(file_path, 'r', encoding='utf-8') as f:  # 确保以 UTF-8 编码读取文件

    data = json.load(f)

# 将数据转换为 JSON 格式的字符串并以漂亮的格式进行打印，并确保输出中文字符
pretty_json = json.dumps(data, indent=4, ensure_ascii=False)
print(pretty_json)

data1 = json.loads(pretty_json)
type(data1)
```

```python
import json

file_path = '/content/drive/My Drive/practice/Tainan.json'

with open(file_path, 'r', encoding='utf-8') as f:  # 确保以 UTF-8 编码读取文件

    data = json.load(f)

# 将数据转换为 JSON 格式的字符串并以漂亮的格式进行打印，并确保输出中文字符
pretty_json = json.dumps(data, indent=4, ensure_ascii=False)
print(pretty_json)

# 將上述結果轉換成dic的形式
data1 = json.loads(pretty_json)

ask = "赤崁樓"
for i in data1:
    if ask == i.get('觀光遊憩區別'):
        print(ask,'的門票收入是',i.get('門票收入'),'元')
```

```python
allage = []
for i in data1:
    allage.append(i.get('上年同月遊客人數'))

allage
```

```python

for i in range(len(allage)):
    allage[i] = str(allage[i]).replace(',', '')

allage_enumerate = enumerate(allage,start = 1)

for index,value in allage_enumerate:
  print(f"Index{index}:{value}")
```
