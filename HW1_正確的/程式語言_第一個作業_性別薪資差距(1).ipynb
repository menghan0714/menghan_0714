```python
data1 = pd.read_csv('/content/drive/My Drive/practice/更新版資料.csv')
```

```python
#從data1中提取欄位名稱為年分、平均工時(女性)和平均工時(男性)的資料

tempset = data1.loc[:, ['年分','平均工時（女性）','平均工時（男性）']]

tempset
```

```python
#將data1中相對應的欄位名稱的數據做指定的數學運算，得出各別年度的性別薪資差距(GOEG)(%)，並另建一個欄位名稱儲存運算結果的資料

data1['性別薪資差距(GOEG)(%)'] = ((((data1['平均工時（男性）'] * data1['時薪（男性）'] * data1['就 業 率（男性）']) - (data1['平均工時（女性）'] * data1['時薪（女性）'] * data1['就業率（女性）']))/(data1['平均工時（男性）'] * data1['時薪（男性）'] * data1['就 業 率（男性）']))*100).apply(lambda x: round(x, 1))
data1
```

```python
#以年分為X軸，平均工時(女性)為Y軸，並利用plotly工具製作出視覺化的折線圖

import plotly.express as px

fig = px.line(tempset, x="年分", y="平均工時（女性）", title='臺灣民國80-111年女性平均工時')
fig.show()
```

```python
#以年分為X軸，平均工時(男性)為Y軸，並利用plotly工具製作出視覺化的折線圖，同時更換折線顏色為紫色

import plotly.express as px

fig = px.line(tempset, x="年分", y="平均工時（男性)", title='臺灣民國80-111年男性平均工時',color_discrete_sequence=['purple'])
fig.show()
```

```python
#以年分為X軸，平均工時(女性)為Y軸，並利用plotly工具製作出視覺化的長條圖，同時更換長條顏色為秘魯色

import plotly.express as px

fig = px.bar(data1, x="年分", y="時薪（男性）", title='臺灣民國80-111年男性時薪',color_discrete_sequence=['peru'])

fig.show()
```

```python
# 因為上圖是用plotly的設定製作出來的，在製圖的一些設定上沒有更改，但我希望能用比較局部的方式去看數據，因此一開始的想法是希望能夠調整y軸的數據差值，讓整個折線圖較能夠呈現不同年分之間的數據差距

import plotly.express as px

fig = px.bar(data1, x="年分", y="時薪（男性）", title='臺灣民國80-111年男性時薪',color_discrete_sequence=['peru'])

fig.update_layout(yaxis=dict(dtick=25))

fig.show()
```

```python
#在嘗試過上述方法後發現不可行，因為他只是更動y軸的數據標記，因此想到可以用省略部分數據(例如:在這裡省略0-100的數據，因為已知我們的數據是介於100-300之間)以及年分資料(80和85年)來讓整個圖能夠呈現近年後局部的數據差異。

import plotly.express as px

fig = px.bar(data1, x="年分", y="時薪（男性）", title='臺灣民國80-111年男性時薪',color_discrete_sequence=['peru'])

fig.update_layout(yaxis=dict(range=[200, 300]))

fig.show()
```

```python
#畫出一個同時包含平均工時(男性)和平均工時(女性)的折線圖，藉此比較男性與女性兩者平均工時的差距以及隨著時代的趨勢

import plotly.express as px


#創建圖表
fig = px.line()

#添加平均工時(男性)的數據
fig.add_scatter(x=data1['年分'], y=data1['平均工時（男性）'], mode='lines', name='男性')

#添加平均工時(女性)的數據
fig.add_scatter(x=data1['年分'], y=data1['平均工時（女性）'], mode='lines', name='女性')

#配置圖表
fig.update_layout(
    title='臺灣男性與女性民國80-111年平均工時趨勢',
    xaxis_title='年分',
    yaxis_title='平均工時'
)

#顯示圖表
fig.show()
```

```python
#畫出一個同時包含時薪(男性)和時薪(女性)的長條圖，藉此比較男性與女性時薪差距以及隨著時代的趨勢

import plotly.express as px

#創建一個包含兩份資料的長條圖
fig = px.bar(data1, x='年分', y=['時薪（男性）', '時薪（女性）'], barmode='group')

#配置圖表
fig.update_layout(
    title='臺灣男性與女性民國80-111年時薪趨勢',
    xaxis_title='年分',
    yaxis_title='時薪'
)

#顯示圖表
fig.show()
```
