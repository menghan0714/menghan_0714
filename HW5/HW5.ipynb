```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
}

url = 'https://zh.wikipedia.org/zh-tw/%E4%BF%84%E4%B9%8C%E6%88%98%E4%BA%89'
response = requests.get(url,headers)
if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')
    paragraphs = soup.find_all('p')
paragraphs
```

```python
text = "\n".join([content.get_text() for content in paragraphs])
text

lines = text.splitlines()[6:]
result_string = '\n'.join(lines)
result_string

import re
cleaned_text = re.sub(r'[A-Za-z!@#$%^&*()_+{}[\]：;"\'<>，.?~\\/\-|=\d（）「」；《》『』\n\xa0“”的和在日月對年與於有及而是亦或為後也被會等了他她卻就雖然著都不又但裏得上]', '', result_string)
cleaned_text = re.sub(r'[\u3000-\u303F]', '', cleaned_text)

cleaned_text

```

```python
import jieba

seg_list = jieba.cut(cleaned_text, cut_all=False)
word_list = list(seg_list)
word_list
```

```python
from gensim import corpora
from gensim.models import LdaModel
from gensim.models.ldamodel import CoherenceModel

tokenized_docs = [jieba.lcut(doc) for doc in word_list]

dictionary = corpora.Dictionary(tokenized_docs)
corpus = [dictionary.doc2bow(doc) for doc in tokenized_docs]

lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=10, passes=10)

topics = lda_model.print_topics(num_words=10)
for topic in topics:
    print(topic)
```
