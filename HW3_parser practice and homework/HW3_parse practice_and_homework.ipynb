```python
import requests
from bs4 import BeautifulSoup

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
}

# 爬取前十页的内容
for page in range(1, 3):
    url = f'https://www.ptt.cc/bbs/Politics/index{page}.html'
    response_test = requests.get(url, headers=headers)

    if response_test.status_code == 200:
        print(f"Page {page}:")

        # 使用 BeautifulSoup 解析 HTML
        soup = BeautifulSoup(response_test.text, 'html.parser')

        # 找到所有帖子的标题和链接
        posts = soup.find_all('div', class_='r-ent')
        for post in posts:
            title = post.find('div', class_='title').text.strip()
            link = 'https://www.ptt.cc' + post.find('a')['href']
            print(f'Title: {title}\nLink: {link}\n')

    else:
        print(f'Failed to fetch page {page}')


```

#第二部分是作業三，網址一樣是ptt的政治板。希望能存取政治板第一頁的文章，並且爬出文章標題、連結、作者和發文時間等資料

```python
import requests
from bs4 import BeautifulSoup
import csv
import json
```

```python
# 設置合適的User-Agent模擬真實用戶瀏覽網站
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.81 Safari/537.36'
}
# 創建能儲存爬蟲結果的列表
post_data = []
```

```python
# 爬取第一頁的内容
for page in range(1,2):
    url = f'https://www.ptt.cc/bbs/Politics/index{page}.html'
    response = requests.get(url, headers=headers)
```

```python
if response.status_code == 200:
    print(f"Page {page}:")

# 使用 BeautifulSoup 解析 HTML
    soup = BeautifulSoup(response.text, 'html.parser')

# 找到第一頁政治板的所有標題和連結
    posts = soup.find_all('div', class_='r-ent')
    for post in posts:
        title = post.find('div', class_='title').text.strip()
        link = 'https://www.ptt.cc' + post.find('a')['href']
        print(f'Title: {title}\nLink: {link}\n')

# 取得訪問每個文章的權限
        post_response = requests.get(link, headers=headers)
        if post_response.status_code == 200:
            post_soup = BeautifulSoup(post_response.text, 'html.parser')

# 將代碼都是<span class="article-metaline">的資料提取出來，會有三組資料(作者、標題、時間)
            metalines = post_soup.find_all('div', class_='article-metaline')

# 從文章內容讀取文章作者以及發文時間等訊息
# 作者項目為第一組資料(metalines[0])，而作者名字則是在<span class="article-meta-value">的這行資料中
# 時間項目為第三組資料(metalines[2])，而時間資料也是在<span class="article-meta-value">的這行資料中
            author = metalines[0].find('span', class_='article-meta-value').text.strip()
            post_time = metalines[2].find('span', class_='article-meta-value').text.strip()
            print(f'Author: {author}\nPost_time: {post_time}\n')

# 將取得的爬蟲資訊整理好
            post_info = {
                'Title': title,
                'Link': link,
                'Author': author,
                'Post_time': post_time

            }
            post_data.append(post_info)  # 將整理分類好的訊息存取到一開始設置的post_data列表

        else:
            print(f'Failed to fetch post {link}')

else:
    print(f'Failed to fetch page {page}')

# 將post_data所整理的資料整併成dataframe
df = pd.DataFrame(post_data)

# 儲存成csv格式
df.to_csv('posts.csv', index=False)

# 儲存成json格式
with open('posts.json', 'w', encoding='utf-8') as json_file:
    json.dump(post_data, json_file, ensure_ascii=False, indent=4)
```
