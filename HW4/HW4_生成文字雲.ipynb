```python
import requests
from bs4 import BeautifulSoup

url = 'https://zh.wikipedia.org/zh-tw/%E5%8F%8D%E5%B0%8D%E9%80%83%E7%8A%AF%E6%A2%9D%E4%BE%8B%E4%BF%AE%E8%A8%82%E8%8D%89%E6%A1%88%E9%81%8B%E5%8B%95'

response = requests.get(url)

if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')

    paragraphs = soup.find_all('p')
paragraphs
```

```python
text = "\n".join([content.get_text() for content in paragraphs])
text

lines = text.splitlines()[10:]
result_string = '\n'.join(lines)
result_string

import re
cleaned_text = re.sub(r'[A-Za-z!@#$%^&*()_+{}[\]：;"\'<>，.?~\\/\-|=\d（）「」；《》『』\n的和在日月對年與於有及而是亦或為後也被會等]', '', result_string)
cleaned_text = re.sub(r'[\u3000-\u303F]', '', cleaned_text)

cleaned_text
```

```python
import jieba
from collections import Counter

seg_list = jieba.cut(cleaned_text, cut_all=False)
word_freq = Counter(seg_list)
word_freq
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

font_path = "/user/share/fonts/truetype/noto/NotoSansCJK-Regular.ttc"
wordcloud = WordCloud(width=800, height=400, font_path=font_path, background_color='white').generate_from_frequencies(word_freq)
```

```python
from google.colab import files

plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.savefig('wordcloud.png')
files.download('wordcloud.png')
plt.show()
```
