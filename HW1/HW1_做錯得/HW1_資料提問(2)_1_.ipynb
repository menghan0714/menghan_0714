```python
# 1. 總共有多少個觀光遊憩景點會需要買門票?

import pandas as pd

# 賦予colab拜訪google drive的權限
from google.colab import drive
drive.mount('/content/drive')

# 存取我的雲端硬碟中practice這個資料夾中檔名為Tainan(2).csv.csv的檔案，並轉為Dataframe的形式
df = pd.read_csv('/content/drive/My Drive/practice/Tainan(2).csv.csv')

# # 列印出Dataframe形式的資料
df
```

```python
# 將第一行臺灣鹽博物館的資料刪除
df_new = df.iloc[1:, :]

# 列印出刪除後的df資料
df_new
```

```python
# 使用counts() 計算備註這一欄中各個值出現的次數
counts = df_new['備註'].value_counts()

# 將備註中為門票數和門票數及人工計數的計算值分別儲存在count1和count2這兩個變數中
count1 = counts.get('門票數', 0)
count2 = counts.get('門票數及人工計數', 0)

# 將各自計算值相加後的結果儲存在total_count變數中
total_count = count1 + count2

# 列印出相加後的結果
print("需要買門票的遊憩景點:", total_count,"個")
```
