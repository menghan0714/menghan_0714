問題1: 想知道在這麼多的臺南觀光景點中有幾個景點示會需要收門票的?

```python
import pandas as pd

# 賦予colab拜訪google drive的權限
from google.colab import drive
drive.mount('/content/drive')

# 存取我的雲端硬碟中practice這個資料夾中檔名為Tainan(2).csv.csv的檔案，並轉為Dataframe的形式
df = pd.read_csv('/content/drive/My Drive/practice/Tainan(2).csv.csv')

# # 列印出Dataframe形式的資料
df
```

```python
# 將第一行臺灣鹽博物館的資料刪除
df_new = df.iloc[1:, :]

# 列印出刪除後的df資料
df_new
```

```python
# 使用counts() 計算備註這一欄中各個值出現的次數
counts = df_new['備註'].value_counts()

# 將備註中為門票數和門票數及人工計數的計算值分別儲存在count1和count2這兩個變數中
count1 = counts.get('門票數', 0)
count2 = counts.get('門票數及人工計數', 0)

# 將各自計算值相加後的結果儲存在total_count變數中
total_count = count1 + count2

# 列印出相加後的結果
print("需要買門票的遊憩景點:", total_count,"個")
```

問題2: 接續前面的問題，在這些會收門票的景點中哪一個景點的票價最高?

```python
# 將門票收入、遊客人次假日、遊客人次非假日這三列具有千位分隔符號的逗號刪除掉
df_new['門票收入'] = df_new['門票收入'].str.replace(',', '')
df_new['遊客人次假日'] = df_new['遊客人次假日'].str.replace(',', '')
df_new['遊客人次非假日'] = df_new['遊客人次非假日'].str.replace(',', '')

# 將門票收入這一列值中含有的金錢符號刪除掉
df_new['門票收入'] = df_new['門票收入'].str.replace('$', '')

# 將門票收入這一列值中沒有資料(顯示NaN)的值替換為0
df_new.fillna(0, inplace=True)

# 列印出處理好的資料
df_new
```

```python
# 將門票收入為0的資料刪除掉
non_zero = df_new[df_new['門票收入'] != 0]

# 將門票收、遊客人次假日、由客人次非假日這三列的資料全部的數據轉換為數值
non_zero['門票收入'] = pd.to_numeric(non_zero['門票收入'])
non_zero['遊客人次假日'] = pd.to_numeric(non_zero['遊客人次假日'])
non_zero['遊客人次非假日'] = pd.to_numeric(non_zero['遊客人次非假日'])

non_zero
```

```python
# 將non_zero資料集中遊客人次假日以及非假日資料中的數值對應相加，並多增加一個欄位(遊客總人次)紀錄運算結果
non_zero['遊客總人次'] = non_zero['遊客人次假日'] + non_zero['遊客人次非假日']

# 將non_zero資料集中門票收入這一列的值除於相對應的遊客總人次，並多增加一個欄位(每張票價)紀錄運算結果
non_zero['每張票價'] = non_zero['門票收入'] / non_zero['遊客總人次']

non_zero
```

```python
non_zero['每張票價'] = non_zero['門票收入'] / non_zero['遊客總人次']

non_zero
```

```python
# 將non_zero資料集中每張票價這一欄位中值最大(票價最貴)的數值儲存在max_price這個變數中

max_price = non_zero['每張票價'].max()

# 將non_zero資料集中每張票價這一欄位中值最大(票價最貴)的數值所對應的資料儲存在max_price_row這個變數中

max_price_row = non_zero[non_zero['每張票價'] == max_price]

max_price_row
```

```python
# 將max_price_row資料集中觀光遊憩區別所對應的資料儲存在result這個變數中
result = max_price_row.loc[12,'觀光遊憩區別']

# 將票價四捨五入至整數部分
rounded_number = round(max_price, 0)

# 將四捨五入後的結果轉換為整數
integer_number = int(rounded_number)

# 列印出來結果
print('票價最高的遊客景點:', result,'，票價為:',integer_number,'元(四捨五入後結果)')
```

問題三: 哪個臺南觀光遊憩景點遊客假日與非假日人次變化率最高?

```python
# 將門票收、遊客人次假日、由客人次非假日這三列的資料全部的數據轉換為數值(避免其中有些數據是字符串未能察覺而造成無法進行數學運算的情況)
df_new['門票收入'] = pd.to_numeric(df_new['門票收入'])
df_new['遊客人次假日'] = pd.to_numeric(df_new['遊客人次假日'])
df_new['遊客人次非假日'] = pd.to_numeric(df_new['遊客人次非假日'])

# 得知遊客假日與非假日人次差距
df_new['遊客假日與非假日人次差距'] = abs(df_new['遊客人次假日'] - df_new['遊客人次非假日'])

# 列印出處理後的資料
df_new
```

```python
# 得知遊客平均人次

df_new['遊客平均人次'] = (df_new['遊客人次假日'] + df_new['遊客人次非假日'])/2

# 將遊客平均人次、遊客假日與非假日人次差距這兩列的資料全部的數據轉換為數值(避免其中有些數據是字符串未能察覺而造成無法進行數學運算的情況)
df_new['遊客平均人次'] = pd.to_numeric(df_new['遊客平均人次'])
df_new['遊客假日與非假日人次差距'] = pd.to_numeric(df_new['遊客假日與非假日人次差距'])

# 得知遊客假日與非假日人次變化率以及轉換為百分率
df_new['遊客假日與非假日人次變化率'] = df_new['遊客假日與非假日人次差距'] / df_new['遊客平均人次']
df_new['遊客假日與非假日人次變化百分率'] = df_new['遊客假日與非假日人次變化率'] * 100

# 列印出處理後的資料
df_new
```

```python
# 將遊客假日與非假日人次變化百分率最大的值取出並儲存在max_percentage這個變數中
max_percentage = df_new['遊客假日與非假日人次變化百分率'].max()

# 將遊客假日與非假日人次變化百分率最大的值所對應的資料儲存在max_percentage這個變數中
max_percentage_row = df_new[df_new['遊客假日與非假日人次變化百分率'] == max_percentage]

# 列印出需要提取的資料
max_percentage_row
```

```python
# 將遊客假日與非假日人次變化百分率最大的觀光遊憩景點提取出來並儲存在result變數中
result = max_percentage_row.loc[15,'觀光遊憩區別']

rounded_ratio = round(max_percentage, 2)

# 列印出提取的資料
print('遊客假日與非假日人次變化的景點:',result,'，變化百分率為: {:.2f}%'.format(rounded_ratio))
```
